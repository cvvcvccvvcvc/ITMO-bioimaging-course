# 00. Extract selected channels

```python
from pathlib import Path
import xml.etree.ElementTree as ET

import pandas as pd
import tifffile

ome_path = Path('data/reg001_expr.ome.tif')
channels_dir = Path('data/channels')
channels_dir.mkdir(parents=True, exist_ok=True)
```

```python
with tifffile.TiffFile(ome_path) as tif:
    root = ET.fromstring(tif.ome_metadata)
    ns = {'ome': root.tag.split('}')[0].strip('{')}
    names = [ch.attrib.get('Name', f'Channel_{i}') for i, ch in enumerate(root.findall('.//ome:Channel', ns))]

channel_table = pd.DataFrame({'index': range(len(names)), 'name': names})
selected_channels = channel_table['name'].tolist()
print(f'Selected {len(selected_channels)} channels')
channel_table
```

```python
name_to_index = {name: i for i, name in enumerate(names)}

with tifffile.TiffFile(ome_path) as tif:
    for name in selected_channels:
        i = name_to_index[name]
        image = tif.pages[i].asarray(maxworkers=4)
        out_path = channels_dir / f'{i:02d}_{name}.tif'
        tifffile.imwrite(out_path, image, compression='deflate')
        print(out_path, image.shape, image.dtype)
```
